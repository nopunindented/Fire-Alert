```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
root_path = r"C:/Users/Matthew/Downloads/"
data_CO2 = ps.read_csv(f"{root_path}ACEFTS_L2_v4p1_CO2.csv")
data_CH3OH = ps.read_csv(f"{root_path}ACEFTS_L2_v4p1_CH3OH.csv")
data_CO = ps.read_csv(f"{root_path}ACEFTS_L2_v4p1_CO.csv")
data_CH4 = ps.read_csv(f"{root_path}ACEFTS_L2_v4p1_CH4.csv")
data_CO2
```

```python
import matplotlib.dates as dates
```

```python
altitude_range = ["0.5","1.5","2.5","3.5","4.5","5.5","6.5","7.5","8.5","9.5"]#,"10.5","11.5","12.5","13.5","14.5","15.5","16.5","17.5","18.5","19.5","20.5","21.5","22.5","23.5","24.5","25.5","26.5","27.5","28.5","29.5","30.5","31.5","32.5"]
start_index = 80000
```

```python
def mean_data(data,altitude_range,start_index):
    return np.nanmean(np.maximum(data[altitude_range][start_index:],0),axis=1)

def max_data(mean_data):
    return np.nanmax(mean_data)

def date_data(data,start_index):
    return data["date"][start_index:].to_numpy()

def pos_data(data,start_index):
    return np.array([data["lat"][start_index:], data["long"][start_index:]],ndmin=2).T

mean_CO2 = mean_data(data_CO2,altitude_range,start_index)
max_CO2 = max_data(mean_CO2)
date_CO2 = date_data(data_CO2,start_index)
pos_CO2 = pos_data(data_CO2,start_index)

mean_CH3OH = mean_data(data_CH3OH,altitude_range,start_index)
max_CH3OH = max_data(mean_CH3OH)
date_CH3OH = date_data(data_CH3OH,start_index)
pos_CH3OH = pos_data(data_CH3OH,start_index)

mean_CO = mean_data(data_CO,altitude_range,start_index)
max_CO = max_data(mean_CO)
date_CO = date_data(data_CO,start_index)
pos_CO = pos_data(data_CO,start_index)

mean_CH4 = mean_data(data_CH4, altitude_range, start_index)
max_CH4 = max_data(mean_CH4)
date_CH4 = date_data(data_CH4,start_index)
pos_CH4 = pos_data(data_CH4,start_index)
```

```python
np.shape(pos_CO2)
pos_CO2[-1]
```

```python
def positive_dev(mean_data,max_data):
    return np.fmax(mean_data/max_data-np.nanmean(mean_data),0)**2

all(date_CO2==date_CH3OH)&all(date_CO==date_CH3OH)&all(date_CH4==date_CH3OH)
```

```python
plt.close()
plt.plot(dates.date2num(date_CO2),mean_CO2/max_CO2,'.')
plt.plot(dates.date2num(date_CH3OH),mean_CH3OH/max_CH3OH,'.')
plt.plot(dates.date2num(date_CO),mean_CO/max_CO,'.')
plt.plot(dates.date2num(date_CH4),mean_CH4/max_CH4,'.')


bad_gas = positive_dev(mean_CO2,max_CO2) + positive_dev(mean_CH3OH,max_CH3OH) + positive_dev(mean_CO,max_CO) + positive_dev(mean_CH4,max_CH4)
plt.plot(dates.date2num(date_CO2[bad_gas>1]),bad_gas[bad_gas>1],'.')
np.c_[date_CO2[bad_gas>1.6],pos_CO2[bad_gas>1.6],bad_gas[bad_gas>1.6]]
```
